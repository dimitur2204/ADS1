```python
class Graph:
    def __init__(self):
        self.vertices = {}
        
    def add_vertex(self, v):
        self.vertices[v] = {}

    def add_edge(self, v1, v2, w):
        self.vertices[v1][v2] = w
        self.vertices[v2][v1] = w
    
    def get_edge(self, v1, v2):
        if v2 in self.vertices[v1]:
            return self.vertices[v1][v2]
        return None
        
    def remove_edge(self, v1, v2):
        del self.vertices[v1][v2]
        del self.vertices[v2][v1]
    
    def remove_vertex(self, rem_v):
        for v in self.vertices:
            if rem_v in self.vertices[v]:
                del self.vertices[v][rem_v]
        del self.vertices[rem_v]
    
    def print_neighbours(self, v):
        for n in self.vertices[v]:
            print( v + " " + str(self.vertices[v][n]) + " <-"  + "> " + n)
        
        
    
g1 = Graph()
g1.add_vertex('A')
g1.add_vertex('B')
g1.add_vertex('C')
g1.add_vertex('D')
g1.add_vertex('E')
g1.add_vertex('F')
g1.add_vertex('G')
g1.add_vertex('H')
g1.add_edge('A','B', 2)
g1.add_edge('A','C', 1)
g1.add_edge('A','D', 7)
g1.add_edge('A', 'E', 9)
g1.add_edge('B', 'E', 1)
g1.add_edge('C', 'F', 6)
g1.add_edge('C', 'G', 12)
g1.add_edge('D', 'E', 3)
g1.add_edge('D', 'G', 2)
g1.add_edge('E', 'G', 8)
g1.add_edge('E', 'H', 7)
g1.add_edge('F', 'G', 2)
g1.add_edge('G', 'H', 1)
g1.print_neighbours('A')
g1.print_neighbours('B')
g1.print_neighbours('C')
g1.print_neighbours('D')
g1.print_neighbours('E')
g1.print_neighbours('F')
g1.print_neighbours('G')
g1.print_neighbours('H')
```
